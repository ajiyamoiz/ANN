```python
# Import libraries
import torch
```

Initialize Values

```python
# Create tensors

# Eta
n = torch.tensor(0.5)

# Inputs
x1 = torch.tensor(0.05)
x2 = torch.tensor(0.10)

# Weights
w1 = torch.tensor(0.15, requires_grad=True)
w2 = torch.tensor(0.20, requires_grad=True)
w3 = torch.tensor(0.25, requires_grad=True)
w4 = torch.tensor(0.30, requires_grad=True)

w5 = torch.tensor(0.40, requires_grad=True)
w6 = torch.tensor(0.45, requires_grad=True)
w7 = torch.tensor(0.50, requires_grad=True)
w8 = torch.tensor(0.55, requires_grad=True)

# Biases
b1 = torch.tensor(0.35, requires_grad=True)
b2 = torch.tensor(0.60, requires_grad=True)

# Losses
l1 = torch.tensor(0.01, requires_grad=True)
l2 = torch.tensor(0.99, requires_grad=True)

x1, x2, w1, w2, w3, w4, w5, w6, w7, w8, b1, b2
```

Forward Pass

```python
# Hidden layer
h1 = 1/(1+torch.exp(-1*(w1*x1+w2*x2+b1*1)))
h2 = 1/(1+torch.exp(-1*(w3*x1+w4*x2+b1*1)))
print(h1)
print(h2)
```

```python
# Output layer
o1 = 1/(1+torch.exp(-1*(w5*h1+w6*h2+b2*1)))
o2 = 1/(1+torch.exp(-1*(w7*h1+w8*h2+b2*1)))
print(o1)
print(o2)
```

```python
# Calculate Error
e1 = torch.sum((l1-o1)**2)/2
e2 = torch.sum((l2-o2)**2)/2
# Total Error
E = e1+e2
print(e1)
print(e2)
print(E)
```

Backward Pass (Back Propagation)

```python
# Compute derivatives
E.backward()
```

```python
# Display gradients
print('dE/dw1:', w1.grad)
print('dE/dw2:', w2.grad)
print('dE/dw3:', w3.grad)
print('dE/dw4:', w4.grad)
print('dE/dw5:', w5.grad)
print('dE/dw6:', w6.grad)
print('dE/dw7:', w7.grad)
print('dE/dw8:', w8.grad)
```

```python
# Update weights
with torch.no_grad():
  w1 = w1-n*w1.grad
  w2 = w2-n*w2.grad
  w3 = w3-n*w3.grad
  w4 = w4-n*w4.grad
  w5 = w5-n*w5.grad
  w6 = w6-n*w6.grad
  w7 = w7-n*w7.grad
  w8 = w8-n*w8.grad

w1, w2, w3, w4, w5, w6, w7, w8
```
